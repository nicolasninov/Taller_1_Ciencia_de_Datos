```python
# Análisis para variables categóricas

variables_categoricas = ['neighbourhood_cleansed', 'room_type', 'latitude', 'longitude']  

print(f"Análisis para la variable categórica '{variables_categoricas}':")
cantidad_elementos_en_dominio_barrio = len(listings_df['neighbourhood_cleansed'].unique())

for variable in variables_categoricas:
    print(f"La variable 'neighbourhood_cleansed' tiene {cantidad_elementos_en_dominio_barrio} posibles barrios.")
    print("Valores Diferentes:")
    print("\nValor Más Común (Moda):")
    print(listings_df[variable].mode()[0])  
    print("\nFrecuencia Absoluta y Relativa:")
    print(listings_df[variable].value_counts()) 
    print(listings_df[variable].value_counts(normalize=True))
    print("\n") 
```

```python
#Análisis variables Cuantitativas

variables_cuantitativas = ['review_scores_rating', 'review_scores_value']

for variable in variables_cuantitativas:
    print(f"Análisis para la variable cuantitativa '{variable}':")
    print("\nModa:")
    print(listings_df[variable].mode()) 
    print("\nMedia:")
    print(listings_df[variable].mean())  
    print("\nMediana:")
    print(listings_df[variable].median())  
    print("\nVarianza:")
    print(listings_df[variable].var())  
    print("\nDesviación Estándar:")
    print(listings_df[variable].std()) 
    print("\n")

    plt.figure(figsize=(8, 6))
    sns.histplot(listings_df[variable], kde=True)
    plt.title(f'Histograma de {variable}')
    plt.xlabel(variable)
    plt.ylabel('Frecuencia')
    plt.show()
```

```python
#LIMPIEZA Y CALIDAD DE DATOS

#Barrios con mínimos valores

barrio_especifico = "la Clota"

conteo_registros = listings_df['neighbourhood_cleansed'].value_counts()

registros_en_barrio = conteo_registros[barrio_especifico]

print(f"El barrio '{barrio_especifico}' tiene {registros_en_barrio} registros.")

#Reseñas para el barrio

barrio_df = listings_df[listings_df['neighbourhood_cleansed'] == barrio_especifico]

total_resenas = len(barrio_df)
promedio_rating = barrio_df['review_scores_rating'].mean()
promedio_value = barrio_df['review_scores_value'].mean()

print(f"Reseñas en el barrio '{barrio_especifico}': {total_resenas}")
print(f"Promedio de review_scores_rating en el barrio: {promedio_rating:.2f}")
print(f"Promedio de review_scores_value en el barrio: {promedio_value:.2f}")

listings_df = listings_df[listings_df['neighbourhood_cleansed'] != barrio_especifico]
```

```python
# CALCULO TASA DE OCUPACION

noches_ocupadas = listings_df['availability_365'].apply(lambda x: 365 - x)  
noches_disponibles = listings_df['availability_365']  

tasa_de_ocupacion = (noches_ocupadas / 365) * 100 

listings_df['tasa_de_ocupacion'] = tasa_de_ocupacion

tasa_de_ocupacion_por_barrio = listings_df.groupby('neighbourhood_cleansed')['tasa_de_ocupacion'].mean()

tasa_de_ocupacion_por_barrio = tasa_de_ocupacion_por_barrio.sort_values(ascending=False)

plt.figure(figsize=(12, 6))
tasa_de_ocupacion_por_barrio.plot(kind='bar', color='skyblue')
plt.title('Tasa de Ocupación por Barrio')
plt.xlabel('Barrio')
plt.ylabel('Tasa de Ocupación (%)')
plt.xticks(rotation=90)
plt.show()

print("Las 5 primeras tasas de ocupación por barrio:")
print(tasa_de_ocupacion_por_barrio.head(5))
```

```python
#MEJORES RESEÑAS POR BARRIO

grouped_by_neighbourhood = listings_df.groupby('neighbourhood_cleansed')


best_reviews = grouped_by_neighbourhood['review_scores_rating'].mean()

best_prices = grouped_by_neighbourhood['review_scores_value'].median()

barrios_mejores_reseñas = best_reviews.sort_values(ascending=False)

barrios_mejores_precios = best_prices.sort_values(ascending=False)

top_5_barrios_mejores_reseñas = barrios_mejores_reseñas.head(5)
top_5_barrios_mejores_precios = barrios_mejores_precios.head(5)

barrios_en_comun = barrios_mejores_reseñas.index.intersection(barrios_mejores_precios.index)

print("Top 5 Barrios con Mejores Reseñas:")
print(top_5_barrios_mejores_reseñas)
print("\nTop 5 Barrios con Mejores Precios:")
print(top_5_barrios_mejores_precios)

primeros_5_barrios_en_comun = barrios_en_comun[:5]
print("\nLos primeros barrios en común con mejor calificación en reseñas en general y precios:")
print(primeros_5_barrios_en_comun)

plt.figure(figsize=(12, 6))
best_reviews.plot(kind='bar', color='skyblue')
plt.title('Mejores Reseñas en General por Barrio')
plt.xlabel('Barrio')
plt.ylabel('Promedio de Review Scores Rating')
plt.xticks(rotation=90)
plt.show()

plt.figure(figsize=(12, 6))
best_prices.plot(kind='bar', color='lightgreen')
plt.title('Mejores Precios por Barrio')
plt.xlabel('Barrio')
plt.ylabel('Mediana de Review Scores Value')
plt.xticks(rotation=90)
plt.show()
```

```python
# RELACIÓN DEL TIPO DE HABITACIÓN Y RESEÑA DE LIMPIEZA EN FUNCIÓN DEL VALOR

plt.figure(figsize=(12, 8))
sns.scatterplot(x='review_scores_cleanliness', y='review_scores_value', data=listings_df, hue='room_type', palette='Blues')
plt.title('Relación entre Review Scores Cleanliness y Review Scores Value (por Room Type)')
plt.xlabel('Review Scores Cleanliness')
plt.ylabel('Review Scores Value')
plt.show()
```

```python
#CALCULO DE RANGO DE PRECIOS DISPUESTOS A PAGAR LOS USUARIOS POR BARRIOS DEPENDIENDO EL TIPO DE HABITACION

listings_df["price_float"] = listings_df["price"].str.replace("[$,]", "", regex=True).astype(float)

q1 = listings_df['price_float'].quantile(0.25)
q3 = listings_df['price_float'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr
filtered_df = listings_df[(listings_df['price_float'] >= lower_bound) & (listings_df['price_float'] <= upper_bound)]

top_neighbourhoods = filtered_df['neighbourhood_cleansed'].value_counts().head(20)
room_types = filtered_df['room_type'].unique()

def calculate_price_range(group):
    q1 = group['price_float'].quantile(0.25)
    q3 = group['price_float'].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    return (group['price_float'][(group['price_float'] >= lower_bound) & (group['price_float'] <= upper_bound)].max() -
            group['price_float'][(group['price_float'] >= lower_bound) & (group['price_float'] <= upper_bound)].min())

price_ranges = {}

for neighbourhood in top_neighbourhoods.index:
    price_ranges[neighbourhood] = {}
    for room_type in room_types:
        group = filtered_df[(filtered_df['neighbourhood_cleansed'] == neighbourhood) & (filtered_df['room_type'] == room_type)]
        price_range = calculate_price_range(group)
        price_ranges[neighbourhood][room_type] = price_range

price_range_df = pd.DataFrame(price_ranges)

price_range_df = price_range_df.T

for room_type in room_types:
    sorted_neighbourhoods = price_range_df[room_type].sort_values(ascending=False)
    top_neighbourhood = sorted_neighbourhoods.index[0]
    top_price_range = sorted_neighbourhoods[0]
    print(f"Barrio con precio más alto para {room_type}: {top_neighbourhood} (Rango de Precios: {top_price_range})")

plt.figure(figsize=(12, 8))
price_range_df.plot(kind='bar', colormap='viridis')
plt.title('Rango de Precios por Barrio y Tipo de Habitación')
plt.xlabel('Barrio')
plt.ylabel('Rango de Precios')
plt.legend(title='Tipo de Habitación')
plt.xticks(rotation=90)
plt.show()
```
